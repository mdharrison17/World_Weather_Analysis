```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
from datetime import datetime
from config import weather_api_key
import requests
from config import g_key
```

```python
# get data from part1 csv
vacation_data_df = pd.read_csv("WeatherPy_vacation")
vacation_data_df.head(10)
```

```python
#filter dataframe to get 4 cities in close proximity
limited_cities = vacation_data_df[vacation_data_df["Country"] == 'AU'].tail(4)
limited_cities
```

```python
#create dataframe for each city
Geraldton_data_df = vacation_data_df[vacation_data_df["City"] == 'Geraldton']
Mareeba_data_df = vacation_data_df[vacation_data_df["City"] == 'Mareeba']
HerveyBay_data_df = vacation_data_df[vacation_data_df["City"] == 'Hervey Bay']
PortHedland_data_df = vacation_data_df[vacation_data_df["City"] == 'Port Hedland']
```

```python
#put lat and long in tuple per city
Geraldton = (Geraldton_data_df[['Lat', 'Lng']].values)
Mareeba = (Mareeba_data_df[['Lat','Lng']].values)
HerveyBay = (HerveyBay_data_df[['Lat', 'Lng']].values)
PortHedland = (PortHedland_data_df[['Lat', 'Lng']].values)
```

```python
#create info for markers
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Weather Description} and {Max Temp} °F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in limited_cities.iterrows()]
```

```python
#lat and long pairs
import gmaps
gmaps.configure(api_key= g_key)
 
#locations coordinates for cities    
locations = limited_cities[["Lat", "Lng"]]

marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

fig = gmaps.figure()
Geraldton2Mareeba = gmaps.directions_layer(Geraldton[0], Mareeba[0], waypoints=[HerveyBay[0],PortHedland[0]], travel_mode = "DRIVING")

fig.add_layer(marker_layer)
fig.add_layer(Geraldton2Mareeba)

fig
```
